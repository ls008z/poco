```python
from simulation.runner import SimulationRunner
from simulation.plotter import SimulationPlotter
from estimators.sample_mean import SampleMeanEstimator
from generators.bootstrap_constant_shifter import BootstrapConstantShifter
from generators.normal_mean_shifter import NormalMeanShifter
import pandas as pd
import numpy as np
```

```python
n_samples = 20
null_value = 0
seed=24
effect_sizes = np.linspace(0,1,40)
n_simulations = 1000
n_jobs = -1
```

```python
# Initialize estimator + generator
estimator = SampleMeanEstimator(column="y", confidence=0.95)
parametric_generator = NormalMeanShifter(n_samples=n_samples, baseline_mean=null_value, std=1)
# Simulated dataset
np.random.seed(seed)
df = pd.DataFrame({"y": np.random.normal(loc=null_value, scale=1, size=n_samples)})
empirical_generator = BootstrapConstantShifter(base_data=df, outcome_col="y")
```

```python

# Run simulations with Parametric Generator
parametric_runner = SimulationRunner(
    estimator=estimator,
    generator=parametric_generator,
    effect_sizes=effect_sizes,
    n_simulations=n_simulations,
    null_value=null_value,
    seed=seed,
    n_jobs=n_jobs,
)
parametric_results = parametric_runner.run()

```

```python
parametric_plotter = SimulationPlotter(parametric_runner)

# Power Curve
fig1 = parametric_plotter.plot_power_curve()
fig1.show()

# Coverage Curve
fig2 = parametric_plotter.plot_coverage_curve()
fig2.show()

# Estimate Distribution
fig3 = parametric_plotter.plot_estimate_distribution(effect_size=effect_sizes[-1])
fig3.show()

# Standard Error Histogram + MCSE overlay
fig4 = parametric_plotter.plot_standard_error_distribution(effect_size=effect_sizes[-1])
fig4.show()

# Bias Curve: mean estimate - true effect
fig5 = parametric_plotter.plot_bias_curve()
fig5.show()

# Standard Error Bias Curve: mean(estimator SE) - std(estimates)
fig6 = parametric_plotter.plot_se_bias_curve()
fig6.show()
```

```python

# Run simulations with Empirical Generator
empirical_runner = SimulationRunner(
    estimator=estimator,
    generator=empirical_generator,
    effect_sizes=effect_sizes,
    n_simulations=n_simulations,
    null_value=null_value,
    seed=seed,
    n_jobs=n_jobs,
)
empirical_results = empirical_runner.run()
```

```python
empirical_plotter = SimulationPlotter(empirical_runner)

# Power Curve
fig1 = empirical_plotter.plot_power_curve()
fig1.show()

# Coverage Curve
fig2 = empirical_plotter.plot_coverage_curve()
fig2.show()

# Estimate Distribution
fig3 = empirical_plotter.plot_estimate_distribution(effect_size=effect_sizes[-1])
fig3.show()

# Standard Error Histogram + MCSE overlay
fig4 = empirical_plotter.plot_standard_error_distribution(effect_size=effect_sizes[-1])
fig4.show()

# Bias Curve: mean estimate - true effect
fig5 = empirical_plotter.plot_bias_curve()
fig5.show()

# Standard Error Bias Curve: mean(estimator SE) - std(estimates)
fig6 = empirical_plotter.plot_se_bias_curve()
fig6.show()
```
